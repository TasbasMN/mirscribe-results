```python
import os

def count_csv_files(root_path):
    csv_count = 0
    
    # Walk through directory and its subdirectories
    for root, dirs, files in os.walk(root_path):
        # Count files with .csv extension
        for file in files:
            if file.lower().endswith('.csv'):
                csv_count += 1
                # Optional: Print each CSV file found
                print(f"Found CSV: {os.path.join(root, file)}")
    
    return csv_count

# Example usage
root_directory = "downloads/560_real"  # Use current directory, or specify your path
total_csvs = count_csv_files(root_directory)
print(f"\nTotal CSV files found: {total_csvs}")
```

```python
import os
import shutil
from pathlib import Path


def copy_csv_files(source_path, destination_path):
    # Create destination directory
    Path(destination_path).mkdir(parents=True, exist_ok=True)
    
    csv_count = 0
    
    # Find and copy all CSV files
    for csv_file in Path(source_path).rglob("*.csv"):
        # Copy file with original name
        destination_file = Path(destination_path) / csv_file.name
        shutil.copy2(csv_file, destination_file)
        csv_count += 1
        print(f"Copied: {csv_file} -> {destination_file}")
    
    return csv_count


# Example usage
source_directory = "downloads/all/for_folder_560_real"  # Source directory (current directory)
destination_directory = "./csv_files"  # Destination directory

total_copied = copy_csv_files(source_directory, destination_directory)
print(f"\nTotal CSV files copied: {total_copied}")
```
