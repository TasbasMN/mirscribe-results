```python
import scipy.stats as stats
import pandas as pd

# Your dataframes
unfiltered_data = real.groupby('mutsig').size()
filtered_data = real[(real.is_gene_upregulated == False) & 
                     (real.is_mirna_oncogene == True)].groupby('mutsig').size()



# Convert to a contingency table manually
contingency_table = pd.DataFrame({
    'unfiltered': unfiltered_data,
    'filtered': filtered_data
})

# Fill any NaN values with 0
contingency_table = contingency_table.fillna(0)

# Perform chi-square test
# chi2_contingency returns: (statistic, p-value, dof, expected)
chi2_result = stats.chi2_contingency(contingency_table)
chi2_stat = chi2_result[0]
p_value = chi2_result[1]

print(f"Chi-square statistic: {chi2_stat}")
print(f"p-value: {p_value}")

# For visualization
# Calculate normalized distributions
unfiltered_normalized = (unfiltered_data / unfiltered_data.sum()) * 100
filtered_normalized = (filtered_data / filtered_data.sum()) * 100

# Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
pd.DataFrame({
    'All Data': unfiltered_normalized,
    'New Oncogenic MREs': filtered_normalized
}).plot(kind='bar')
plt.title('Normalized Distribution Comparison')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate percentage change
unfiltered_normalized = (unfiltered_data / unfiltered_data.sum()) * 100
filtered_normalized = (filtered_data / filtered_data.sum()) * 100
pct_change = ((filtered_normalized - unfiltered_normalized) / unfiltered_normalized) * 100

# Create a comparison plot
plt.figure(figsize=(15, 8))

# Create color list based on values
colors = ['green' if x >= 0 else 'red' for x in pct_change]

plt.bar(range(len(pct_change)), pct_change, color=colors)
plt.xticks(range(len(pct_change)), pct_change.index, rotation=45)
plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
plt.title('Percentage Change in Signature Distribution After Filtering')
plt.ylabel('Percentage Change (%)')
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt

# Your three datasets
unfiltered_data = real.groupby('mutsig').size()
filtered_data1 = real[(real.is_gene_upregulated == False) & 
                     (real.is_mirna_tumor_suppressor == True)].groupby('mutsig').size()
filtered_data2 = real[(real.is_gene_upregulated == True) & 
                     (real.is_mirna_tumor_suppressor == True)].groupby('mutsig').size()

# Contingency table and chi-square for filter2
contingency_table2 = pd.DataFrame({
    'unfiltered': unfiltered_data,
    'filtered': filtered_data2
}).fillna(0)

chi2_result2 = stats.chi2_contingency(contingency_table2)
print("Filter 2 (upregulated=True):")
print(f"Chi-square statistic: {chi2_result2[0]}")
print(f"p-value: {chi2_result2[1]}")

# Calculate normalized distributions
unfiltered_normalized = (unfiltered_data / unfiltered_data.sum()) * 100
filtered_normalized2 = (filtered_data2 / filtered_data2.sum()) * 100

# Calculate percentage change
pct_change2 = ((filtered_normalized2 - unfiltered_normalized) / unfiltered_normalized) * 100

# Create a comparison plot
plt.figure(figsize=(15, 8))
colors = ['green' if x >= 0 else 'red' for x in pct_change2]

plt.bar(range(len(pct_change2)), pct_change2, color=colors)
plt.xticks(range(len(pct_change2)), pct_change2.index, rotation=45)
plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
plt.title('Percentage Change in Signature Distribution (Upregulated=True)')
plt.ylabel('Percentage Change (%)')
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Print top changes
print("\nBiggest changes in signatures (Upregulated=True):")
print(pct_change2.sort_values(ascending=False))
```

```python
# Calculate normalized distributions for all three
unfiltered_normalized = (unfiltered_data / unfiltered_data.sum()) * 100
filtered_normalized1 = (filtered_data1 / filtered_data1.sum()) * 100
filtered_normalized2 = (filtered_data2 / filtered_data2.sum()) * 100

# Create DataFrame for plotting
comparison_df = pd.DataFrame({
    'Unfiltered': unfiltered_normalized,
    'Downregulated': filtered_normalized1,
    'Upregulated': filtered_normalized2
})

# Plot
plt.figure(figsize=(15, 8))
comparison_df.plot(kind='bar', width=0.8)
plt.title('Mutational Signature Distribution Comparison')
plt.xlabel('Signatures')
plt.ylabel('Percentage (%)')
plt.legend(title='Filters')
plt.grid(True, axis='y', alpha=0.3)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Print the actual values
print("\nNormalized distributions:")
print(comparison_df)
```
